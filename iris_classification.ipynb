# Sigmoid Function
Sigmoid function is commonly used in logistic regression and artificial neurons. It is used to convert the continious values into the more of binary values. In neural networks its also called as activation function.

```python
def Sigmoid(val):
  return 1/(1+math.e**(-val)) # fromula for sigmoid function

# for example this makes a S-curve

x = [0.1*i for i in range(-25,26)]
y = [Sigmoid(i) for i in x]
plt.plot(x,y)
```

# Lost Function
Lost function is used to compare the predicted output with the expected output

```python
def loss(y_train,y_predicted):
  return -(y_train * math.log(y_predicted) + (1-y_train) * math.log((1-y_predicted)))
```
